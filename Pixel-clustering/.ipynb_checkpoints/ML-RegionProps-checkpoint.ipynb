### This is the function to read tiff images. Mind the 'folder' parameter to set whether it is a multiframe tiff or multiple tiff images.

```python
def napari_read_tiff(pathlibpath,start=0, folder=False, nframes='all'):
    '''
    Read a multiframe tif image and loads it in a numpy array
    '''
    if not folder:
        with Image.open(str(pathlibpath), mode='r') as img:
            stack = []
            if nframes == 'all':

                try:
                    for i in range(start,img.n_frames):
                        img.seek(i)
                        stack.append(np.array(img))
                except:
                    print(i)
                    pass
        #     else:
        #         for i in range(start,nframes):
        #             img.seek(i)
        #             stack.append(np.array(img))
    if folder:
        files = listdir(pathlibpath)
        stack = []
        for file in files:
            filepath = pathlibpath / file
            with Image.open(str(filepath),mode='r') as img:
                stack.append(np.array(img))
    
    return np.transpose(np.array(stack), (1, 2, 0))
```

### Napari has different coordinate system than ImageJ. I use this func to make it equivalent to imagej

```python
def axis_napari(array):
    array = np.transpose(array, (2,0,1))
    return(array)
```

### Functions

```python
def midbscan(array,max_dist=70, min_samples=1):
    a_coord = np.argwhere(array>0)
    neigh = NearestNeighbors(n_neighbors=100)
    nbrs = neigh.fit(a_coord)
    distances, indices = nbrs.kneighbors(a_coord)
    m = DBSCAN(eps=max_dist, min_samples=min_samples)
    m.fit(a_coord)
    clusters = m.labels_
    return clusters,a_coord
```

```python
def get_img_clusters(array,clusters,a_coord):
    img1_dummy = np.zeros((array.shape))
    l_puntos = [item for item in zip(a_coord[:,0],a_coord[:,1],clusters)]
    for row in l_puntos:
        img1_dummy[row[0],row[1]] = row[2]
    return img1_dummy
```

```python
imgs = [img2,img3,img4]
results = []
for imagen in imgs:
    results.append(midbscan(imagen,max_dist=55))
# l_clusters = results
l_clusters = [item[0] for item in results]
l_a_coords = [item[1] for item in results]
l_clusters,l_a_coords
```

```python
viewer.add_image(get_img_clusters(img2,l_clusters[0],l_a_coords[0]))
```
